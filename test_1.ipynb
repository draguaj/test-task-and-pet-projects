### Предлагаю категоризировать звонки

Классифицируем длительность (Conver_minute) общения с оператором, скорость решения проблемы. Попробуем понять, есть ли классы задач, которые отнимают очень много времени.
1.	Нет контакта (обратный звонок без ответа, Сорвался/тишина и т.д. ,длительность 0, сервис перезвонил, сервис не перезвонил)
2.	До 60 сек
3.	60 – 179
4.	180 – 299
5.	300  и более


В приложении JMP бегло просмотрел названия тем звонков, по которым контакта не состоялось. Чтобы не ошибиться копируем темы, предварительно выводя уникальные значения для просмотра.

```python
tuple(sorted(data.INTERACTION_REASON_LEVEL_2.unique()))
```

```python
data.loc[data.CALL_DURATION < 60, 'CONVER_MINUTE'] = '2. < 60'
data.loc[(data.CALL_DURATION >= 60) & (data.CALL_DURATION < 180), 'CONVER_MINUTE'] = '3. 60-179'
data.loc[(data.CALL_DURATION >= 180) & (data.CALL_DURATION < 300), 'CONVER_MINUTE'] = '4. 180-299'
data.loc[data.CALL_DURATION >= 300, 'CONVER_MINUTE'] = '5. >= 300'
data.loc[data.CALL_DURATION == 0, 'CONVER_MINUTE'] = '1. no contact'
data.loc[data.INTERACTION_REASON_LEVEL_2.str.contains('Сорвался'), 'CONVER_MINUTE'] = '1. no contact'
data.loc[data.INTERACTION_REASON_LEVEL_3.str.contains('Клиент не берет трубку|Сервис не перезвонил|Сервис перезвонил, но не произошло соединение с оператором'), 'CONVER_MINUTE'] = '1. no contact'
```

Категориальные переменные переводим в integer (целочисленные) путем кодирования.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()
le5 = LabelEncoder()
le6 = LabelEncoder()
le7 = LabelEncoder()
le8 = LabelEncoder()
le9 = LabelEncoder()
le10 = LabelEncoder()
le11 = LabelEncoder()
le12 = LabelEncoder()
le13 = LabelEncoder()
```

```python
le.fit(data_LR['SEGMENT'])
le2.fit(data_LR['REGION'])
le3.fit(data_LR['INTERACTION_REASON_LEVEL_1'])
le4.fit(data_LR['INTERACTION_REASON_LEVEL_2'])
le5.fit(data_LR['INTERACTION_REASON_LEVEL_3'])
le6.fit(data_LR['INTERACTION_REASON_LEVEL_4'])
le7.fit(data_LR['DEVICE'])
le8.fit(data_LR['ROAM_6MTH'])
le9.fit(data_LR['ARPU_VOICE'])
le10.fit(data_LR['ARPU_DATA'])
le11.fit(data_LR['ROAM_ON_CALL'])
le12.fit(data_LR['DELTA_TIME_CATEGORY'])
le13.fit(data_LR['CONVER_MINUTE'])
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
